Teste de hipóteses

* $H_0$: θ ∈ $Θ_{0}$ hipótese de nulidade
* $H_1$: θ ∈ $Θ_1$  hipótese alternativa

Tipos de erros
\begin{array}{|cc|} \hline
    Decisão \\ \hline
 &  Rejeita H_0 &  Não Rejeita H_0 \\ \hline
H_0 é V & Erro Tipo I & Acerto \\
H_0 é F & Acerto & Erro Tipo II \\ \hline
\end{array}

Poder do Teste ($π=1-β$), onde \pi é a probabilidade de rejeitar $H_0$ dado que $H_0$ é falsa

Etapas de um teste de hipóteses

* Definir a hipótese nula ($H_0$) e a hipótese alternativa ($H_1$)
* Escolher a estatística adequada para julgar a hipótese nula ($H_0$)
* Fixar o nível de significância do teste e definir a região crítica 
* Usar as informações da amostra para encontrar o valor observado da estatística
* Concluir

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(1)
```

```python
x = np.random.normal(loc=3, scale=2, size=10)
plt.hist(x, bins=30);
```

```python
stats.ttest_1samp(x, 3)  # alternative='two-sided'
```
